```python
sales_data = pd.concat([df1, df2], ignore_index=True)

# Checking IF Data is Scattered
print("Is Order ID unique =", sales_data['Order_ID'].is_unique)

print( "Unique Order ID =", sales_data['Order_ID'].nunique())
```

```python
# Converting Date to datetime format
sales_data['Date'] = pd.to_datetime(sales_data['Date'])

sales_data = sales_data.sort_values(by='Date', ascending=False)

sales_data.head(10)
```

```python
profile = ProfileReport(sales_data, title='raw_data', explorative=True)
profile.to_file("revenue_analysis_raw_data.html", silent=True)
```
