```python
# Dropping the columns that are not relevant for the analysis and lots of NAN values in the dataset

df_clean = df.drop(
    columns=[ 'domain','involvement.Alcoholic Beverages', 'involvement.Adult Entertainment', 'involvement.Gambling', 'involvement.Tobacco Products', 'involvement.Animal Testing', 'involvement.Fur and Specialty Leather', 'involvement.Controversial Weapons', 'involvement.Small Arms', 'involvement.Catholic Values', 'involvement.GMO', 'involvement.Military Contracting', 'involvement.Pesticides', 'involvement.Thermal Coal', 'involvement.Palm Oil', 'employees', 'industry', 'sector', 'sdg.No Poverty', 'sdg.No Hunger', 'sdg.Good Health and Well-Being', 'sdg.Quality Education', 'sdg.Gender Equality', 'sdg.Clean Water and Sanitation', 'sdg.Affordable and Clean Energy', 'sdg.Decent Work and Economic Growth', 'sdg.Industry, Innovation and Infrastructure', 'sdg.Reduced Inequalities', 'sdg.Sustainable Cities and Communities', 'sdg.Responsible Consumption and Production', 'sdg.Climate Action', 'sdg.Life under Water', 'sdg.Life on Land', 'sdg.Peace, Justice and Strong Institutions', 'sdg.Partnerships for the Goals', 'Controversies.Supply Chain Labor Standards', 'Controversies', 'Decarbonization Target', 'sdg', 'Controversies.Collective Bargaining & Union', 'Controversies.Health & Safety', 'Controversies.Discrimination & Workforce Diversity', 'Controversies.Labor Management Relations', 'involvement', 'involvement_msci', 'Controversies.Anticompetitive Practices', 'Controversies.Privacy & Data Security', 'Controversies.Bribery & Fraud', 'Controversies.Governance Structures', 'Controversies.Customer Relations', 'Controversies.Product Safety & Quality', 'Controversies.Human Rights Concerns', 'Controversies.Energy & Climate Change', 'Controversies.Toxic Emissions & Waste', 'Controversies.Impact on Local Communities', 'Controversies.Biodiversity & Land Use', 'Controversies.Other', 'Controversies.Marketing & Advertising', 'Controversies.Civil Liberties', 'Controversies.Operational Waste (Non-Hazardous)', 'Controversies.Supply Chain Management', 'Controversies.Water Stress', 'Controversies.Child Labor', 'Controversies.Controversial Investments' ],
    inplace=False
    )
```

```python
df_clean.info()
```

```python
df_clean.head()
```

```python
# Removing the errors values in datasets

cols = [
    'Decarbonization Target.Comprehensiveness',
    'Decarbonization Target.Ambition p.a.'
]

for col in cols:
    df_clean[col] = (
        df_clean[col]
        .astype(str)
        .str.replace(r'\((.*?)\)', r'-\1', regex=True)  # handle (123) negatives
        .str.replace(r'[^\d\.-]', '', regex=True)       # remove %, //, text, etc
    )
    
    df_clean[col] = pd.to_numeric(df_clean[col], errors='coerce')
```

```python
# Checking Data Quality

obj_cols = df_clean.select_dtypes(include='object').columns


for col in obj_cols:
    print(f"\n===== {col} =====")
    print(df_clean[col].unique()[:10])
```

```python
# Filling the NaN values in the 'Decarbonization Target.Comprehensiveness' and 'Decarbonization Target.Ambition p.a.' columns with 0, assuming that NaN indicates no target set
cols = [
    'Decarbonization Target.Decarbonization Target',
    'Decarbonization Target.Decarbonization Target on Temperature Rise'
]

df_clean[cols] = df_clean[cols].fillna('No')

print(df_clean['Decarbonization Target.Decarbonization Target'].unique())
print(df_clean['Decarbonization Target.Decarbonization Target on Temperature Rise'].unique())
```

# Feature Engineering

```python
# Encoding controversy columns
controversies_map = {
    'Green': 0,
    'Yellow': 1,
    'Orange': 2,
    'Red': 3,
}

columns = [
    'Controversies.Environment', 'Controversies.Social', 'Controversies.Customers', 'Controversies.Human Rights & Community', 'Controversies.Labor Rights & Supply Chain', 'Controversies.Governance'
]

for tag in columns:
    df_clean[tag] = df_clean[tag].map(controversies_map)

# Since the controversy columns have lots of missing values, we can sum them up to get a single score
df_clean['Controversies_score_ESG'] = df_clean[columns].sum(axis=1)
```

```python
# Encoding binalry columns
binary_cols = [
    'Decarbonization Target.Decarbonization Target',
    'Decarbonization Target.Decarbonization Target on Temperature Rise',
    'involvement_msci.Controversial Weapons',
    'involvement_msci.Gambling',
    'involvement_msci.Tobacco Products',
    'involvement_msci.Alcoholic Beverages'
]

df_clean[binary_cols] = df_clean[binary_cols].apply(lambda x: x.map({'Yes': 1, 'No': 0})).fillna(0)
```

```python
df_clean.describe()
```

```python
# Key Features missing values
df_clean[['Decarbonization Target.Target Year', 'Decarbonization Target.Comprehensiveness', 'Decarbonization Target.Ambition p.a.', 'Temperature Goal', 'esg']].isna().sum()
```

```python
# Due large amount of missing values, converted disclosed information into binary variables

# Aligned with Global target (Paris Agreement)
df_clean['global_target_aligned'] = (df_clean['Decarbonization Target.Target Year'] <= 2050).astype(int) 

# Aligned with SBTI long-term reduction target
df_clean['SBTI_reduction_aligned'] = (df_clean['Decarbonization Target.Comprehensiveness'] <= 90).astype(int) 

# Aligned with Global temperature target
df_clean['Tempurature_aligned'] = (df_clean['Temperature Goal'] <= 1.5).astype(int) 
```

```python
# Dropping the columns that are not relevant for the analysis and lots of NAN values in the dataset, ESG column has been dropped since we don't have a information how the score got derived and lots of missing values

model_data = df_clean.drop(
    columns=['name', 'ticker', 'Decarbonization Target.Decarbonization Target on Temperature Rise', 'Decarbonization Target.Decarbonization Target' , 'Decarbonization Target.Target Year', 'Decarbonization Target.Comprehensiveness', 'Decarbonization Target.Ambition p.a.', 'Temperature Goal', 'esg', 'Controversies.Environment', 'Controversies.Social', 'Controversies.Customers', 'Controversies.Human Rights & Community', 'Controversies.Labor Rights & Supply Chain', 'Controversies.Governance' ],
    axis=1
).reset_index(drop=True)
```
